## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd
import numpy as np
import os
import subprocess

filename = "content_refresh_anonymized.csv"
df = None

# Search dynamically for the file anywhere in the current working directory or system
for root, dirs, files in os.walk("."):
    if filename in files:
        file_path = os.path.join(root, filename)
        df = pd.read_csv(file_path)
        print(f"Successfully loaded data from dynamic path: {file_path}")
        break

# Fallback: If the repository isn't cloned right or file is missing, pull it directly via raw git URL
if df is None:
    print(f"File '{filename}' not found locally. Pulling fallback dataset from raw starter repository...")
    raw_url = "https://raw.githubusercontent.com/flyrank-bih/flyrank-ml-internship-starter/main/data/raw/content_refresh_anonymized.csv"
    try:
        df = pd.read_csv(raw_url)
        print("Successfully loaded fallback dataset from remote server!")
    except Exception as e:
        raise FileNotFoundError(f"Could not locate or pull {filename}. Error details: {e}")

print("\n--- LIVE REPO DATA INSIGHTS ---")
# 1. Total volume of unique pages analyzed
print(f"Total Portfolio Pages Analysed: {df.shape[0]:,}")

# 2. Baseline Correlation between volume and actual traffic
volume_traffic_corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between Search Volume & Realized Impressions: {volume_traffic_corr:.4f}")

# 3. Absolute breakdown of pages currently on a downward trend
decay_count = (df["trend_direction"] == "down").sum()
decay_pct = (df["trend_direction"] == "down").mean() * 100
print(f"Pages with Active 'down' Trend: {decay_count:,} ({decay_pct:.2f}%)")
```
