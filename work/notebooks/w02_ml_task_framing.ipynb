## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
import pandas as pd
import numpy as np
import os

# Bulletproof dataset recovery routine
filename = "content_refresh_anonymized.csv"
df = None
for root, dirs, files in os.walk("."):
    if filename in files:
        df = pd.read_csv(os.path.join(root, filename))
        break

if df is None:
    raw_url = "https://raw.githubusercontent.com/flyrank-bih/flyrank-ml-internship-starter/main/data/raw/content_refresh_anonymized.csv"
    df = pd.read_csv(raw_url)

print("=== DEFINING THE UNIT OF ANALYSIS ===")
print("One single row = One unique anonymized URL entity asset evaluated over a 90-day performance snapshot.\n")

# Dynamically pick available columns to prevent KeyError
available_cols = list(df.columns)
target_candidates = ['trend_direction', 'trend', 'direction']
feature_candidates = ['avg_position', 'ctr', 'word_count', 'search_volume', 'impressions_90d', 'content_age_days']

# Find which columns actually exist in your dataset
selected_features = [col for col in feature_candidates if col in available_cols]
actual_target = [col for col in target_candidates if col in available_cols]

# Use the first column as the unique asset identifier if a specific ID column isn't found
id_col = [col for col in ['page_id', 'id', 'url_id'] if col in available_cols]
if not id_col:
    id_col = [available_cols[0]]  # Fallback to the first column (usually the index or identifier)

final_cols_to_show = id_col + selected_features + actual_target

# Build and display the clean slice
df_slice = df[final_cols_to_show].copy()

# Sketch the binary target column live based on whatever the trend column is called
main_target_col = actual_target[0] if actual_target else None
if main_target_col:
    df_slice['is_decaying_target'] = (df_slice[main_target_col] == 'down').astype(int)

print("Dataframe Sample:")
print(df_slice.head(5).to_string(index=False))
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
# Section 5: The Algorithmic Edge Over Rules
ml_vs_rules = (
    "A simple rule-based approach like 'if a page is over 365 days old and CTR dropped, flag it' "
    "fails to account for non-linear feature interaction. For example, a drop in CTR might be perfectly "
    "normal if a page drops from rank 1 to rank 3 due to seasonality, but highly critical if it happens "
    "at rank 15. A machine learning model (like a Random Forest or Decision Tree) captures these multi-variable "
    "conditional dependencies concurrently, outperforming standard static heuristic filters."
)

print(f"Why ML is Mandatory: {ml_vs_rules}")
```
